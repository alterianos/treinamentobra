### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```
import numpy as np

arr = np.random.rand(50)

print(np.average(arr))
print(np.std(arr))
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
arra = np.random.rand(10)

print(arra)
for arra in arra:
  if arra <= 0.5:
    arra = arra
    print(np.sqrt(arra))
  
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
import numpy as np

matriz_3d = np.random.rand(3,3,3)

max_linha2 = np.amax(matriz_3d, axis=1)

print("Valor de cada linha \n", matriz_3d)
print("Valor máximo de cada linha",max_linha2)
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```
import numpy as np


#axis = 1 --> Linha
#axis = 0 --> Coluna
matriz_5 = np.random.rand(5,5)
 
media = np.average(matriz_5, axis = 1)
sub_media = matriz_5 - media[:, np.newaxis]

print("Matriz", matriz_5)
print("Média de cada linha", media)
print("Subtração da média de cada linha", sub_media)

```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
import numpy as np 
 
matriz_4 = np.random.rand(4,4)

diagonais_matriz = np.diag(matriz_4)
media_matriz4 = np.average(diagonais_matriz)
desvio_matriz4 = np.std(diagonais_matriz)

print("Matriz 4x4 \n",matriz_4)
print("Diagonais da matriz \n", diagonais_matriz)
print("Media das diagonais da matriz \n", media_matriz4)
print("Desvio padrão das diagonais da matriz \n ", desvio_matriz4)

```

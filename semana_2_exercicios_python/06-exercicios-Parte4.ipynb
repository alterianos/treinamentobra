### 21. Crie um programa em Python que receba uma string como entrada e imprima a string sem as vogais.

```python
def remove_vogal():
  string = input("Digite uma string: ")
  vogais=['a','e','i','o','u','A','E','I','O','U']
  for i in range(len(vogais)):
    string =string.replace(vogais[i],'')
  return string

remove_vogal()
```

### 23. Crie um programa em Python que receba uma lista de números e imprima a soma dos números pares e a soma dos números ímpares presentes na lista.

```python
def pares_e_impares(lista):
  soma_par=0
  soma_impar=0
  for elem in lista:
    if elem%2==0:
      soma_par+=elem
    else: 
      soma_impar+=elem
  print(f'soma dos elementos pares: {soma_par}\nsoma dos elementos impares: {soma_impar}')
  return

pares_e_impares([1,2,3,4,5,6,7,8])
```

### 24. Crie um programa em Python que receba uma lista de strings e retorne uma lista com todas as strings que começam e terminam com a mesma letra.

```python
palavras = ["amor", "futebol", "verdade", "banana", "arara", "casa", "sol"]
def acabo_a_criatividade_2(lista):
  palavras_selecionadas = []
  for elem in lista:
    if elem[0]==elem[-1]:
      palavras_selecionadas.append(elem)
  return palavras_selecionadas

acabo_a_criatividade_2(palavras)
```

### 25. Crie um programa em Python que leia um número do usuário e verifique se ele é um número perfeito. Um número perfeito é aquele que é igual à soma de seus divisores próprios (ou seja, a soma de seus fatores inteiros positivos, excluindo ele próprio).

```python
def eh_perfeito():
  numero = int(input("Digite um número: "))
  soma=0
  for i in range(1,numero):
    if numero%i==0:
      soma+=i
  if soma==numero:
    return True
  return False

eh_perfeito()
```

### 1. Some todos os números de 1 até 100

```python
soma = 0
for i in range(101):
  soma += i
  print(soma)
```

### 2. Some todos os números pares de 1 até 100

```python
soma = 0
for n in range(0,101):
  if n % 2 == 0:
    soma += n
    print(soma)
  
```

### 3. Some todos os números ímpares de 1 até 100

```python
soma = 0
for n in range(0,100):
  if n % 2 > 0:
    soma += n
    print(soma)
```
